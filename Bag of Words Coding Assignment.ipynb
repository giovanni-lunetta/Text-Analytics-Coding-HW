```python
import re

import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import nltk
```

```python
with open("Peter.txt") as file:
   peter_string = file.read()
with open("Eric.txt") as file:
   eric_string = file.read()
print(eric_string)
```

```python
def split_and_clean_transcript(string_transcript):
    list_transcript = string_transcript.split("\n")
    clean_transcript = list()
    for item in list_transcript:
        if item.startswith('*') == True:
            clean_utterance = item.replace("\x15", "")
            clean_utterance = clean_utterance.replace("\t", " ")
            clean_utterance = re.sub("[0-9]*_[Z0-9]*", "", clean_utterance)
            clean_utterance = re.sub("\[(.*?)\]", " ", clean_utterance)
            clean_transcript.append(clean_utterance)
    return clean_transcript
```

```python
peter_transcript = split_and_clean_transcript(peter_string)
eric_transcript = split_and_clean_transcript(eric_string)
eric_transcript[0:50]
```

```python
df_peter = pd.DataFrame()
df_peter["text_raw"] = peter_transcript
df_peter["transcript"] = "peter" 
df_peter
```

```python
df_eric = pd.DataFrame()
df_eric["text_raw"] = eric_transcript
df_eric["transcript"] = "eric"
df_eric
```

```python
df = pd.concat([df_eric, df_peter])
df.sample(10)
```

```python
df.info()
```

```python
# A speaker tag is a substring within a transcript that indicates who is speaking. 
# In these transcripts, the first five characters of every row in text_raw column of the dataframe is a speaker tag. 
# Create a column named "speaker_tag" that contains the first five characters of "text_raw". 
# Create a column named "text" that contains every character after the fifth character of "text_raw". Provide the code below.

df["speaker_tag"] = df["text_raw"].str[0:5]
df["text"] = df["text_raw"].str[5:]

df
```

```python
# Recode the values in speaker using the schema provided in the table. 
# So, when the value in speaker tag is “*CHI:” it should now be child. Provide the code below.

df["speaker"] = df["speaker_tag"].replace({"*CHI:": "child", "*PAT:": "researcher1", "*LOI:": "researcher2", "*MOT:": "mother"})

df
```

```python
# Now, create a new column called speaker_full that contains the text in transcript column, followed by an underscore, followed by the text in speaker. 
# So, the first row of the dataframe with contain “eric_researcher2” in the speaker_full column. Provide the code below.

df["speaker_full"] = df["transcript"] + "_" + df["speaker"]

df
```

```python
# Who speaks more, Peter or Eric? Provide the code below.

df.speaker_full.value_counts()
```

```python
# Make all of the text in text lowercase. Provide the code below.

df["text_lower"] = df["text"].str.lower()

df
```

```python
# Use the following code to create df_speakers. Explain what the code does by explaining every method in the code.

df_speakers = df.groupby(["transcript", "speaker", "speaker_full"])["text"].apply(' '.join).reset_index()

df_speakers
```

```python
# Set the index to be the speaker_full column. Provide the code below.

df_speakers = df_speakers.set_index("speaker_full")

df_speakers
```

```python
# Create a new column in df_speakers that contains the number of times the speaker speaks. Provide the code below. (Note: This question is meant to be a bit of a challenge.)

df_speakers["num_utterances"] = df_speakers["text"].apply(lambda x: len(x.split()))

df_speakers["num_utterances"]
```

```python
speaker_counts = df['speaker'].value_counts().reset_index()
speaker_counts.columns = ['speaker', 'num_speaks']

# Left join with the original DataFrame
df_speakers = df_speakers.merge(speaker_counts, on='speaker', how='left')
```
